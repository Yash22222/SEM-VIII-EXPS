```python
G = nx.Graph()
users = ["A", "B", "C", "D", "E", "F", "G", "H"]
G.add_nodes_from(users)
edges = [("A", "B"), ("A", "C"), ("B", "D"), ("C", "D"), ("E", "F"), ("F", "G"), ("G", "H"), ("E", "H")]
G.add_edges_from(edges)
partition = community.best_partition(G)
colors = [partition[node] for node in G.nodes()]
plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color=colors, cmap=plt.cm.jet, node_size=1000, edge_color='gray')
plt.title("Community Detection in a Social Network")
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_centrality)
pagerank = nx.pagerank(G, alpha=0.85)
print("PageRank:", pagerank)
most_influential = max(pagerank, key=pagerank.get)
print(f"The most influential user is: {most_influential}")
```
