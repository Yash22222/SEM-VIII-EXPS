```python
import pandas as pd
import re
from IPython.display import display
```

```python
df = pd.read_csv('reddit_comments.csv')
df.to_csv('raw_reddit_comments.csv', index=False)
```

```python
print(df.head)
```

```python
df['Comment Author'].fillna("Unknown", inplace=True)
df['Comment'].fillna("No Comment", inplace=True)

def clean_text(text):
    if isinstance(text, str): # Check if it's a string
        text = re.sub(r'http\S+', '', text)
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        text = text.lower()  # Convert to lowercase
        return text
    else:
      return "" # or handle non-string values differently

df['Comment'] = df['Comment'].apply(clean_text)

df.drop_duplicates(subset=['Comment'], inplace=True)

df_filtered = df[df['Score'] > 5]  # Adjust filter as needed

df_filtered.to_csv('cleaned_reddit_comments.csv', index=False)

print("Data cleaning and storage complete!")
```
